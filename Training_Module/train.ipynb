# DATA ACQUISITATION E PREPROCESS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
FILENAME = "train.csv"
df = pd.read_csv(FILENAME)

```

```python
num_rows, num_cols = df.shape
print("Number of rows: ", num_rows)
print("Number of columns: ", num_cols)
```

Non dovremmo avere valori nulli

```python
df.info()
```

Controllo presenza e cancello eventuali duplicati 

```python
duplicates = df.duplicated()
print("Number of duplicate rows: ", duplicates.sum())

df = df.drop_duplicates()
df
```

Controllo distribuzione anno di pubblicazione

```python
plt.hist(df['Year'])
plt.xlabel('Anno di pubblicazione')
plt.ylabel('Frequenza')
plt.title('Distribuzione anno di pubblicazione')
plt.show()
```

Matrice di correlazione

```python
matrix_corr = df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(matrix_corr, annot=True, cmap='coolwarm', fmt=".2f",linewidths=.5, xticklabels=False, yticklabels=False, cbar=False)
plt.title('Matrice di Correlazione')
plt.show()
```

```python
Y = df.iloc[:, 0]
X = df.iloc[:, 1:]
```

```python
from sklearn.preprocessing import MinMaxScaler
import pickle

file = open("scaler.save","wb")
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
print(X)
pickle.dump(scaler, file)
```

```python
#PCA 
from sklearn import decomposition
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

pca = decomposition.PCA()
pca.fit(X_train)

print(X_train)


X_train_t = pca.transform(X_train)
plt.scatter(X_train_t[:, 0], X_train_t[:, 1], c=y_train)
plt.show()


print(X_train_t)
```

# MODELLI

### LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


regr = LinearRegression()
regr.fit(X_train, y_train)

# Effettua previsioni sui dati di test
y_pred = regr.predict(X_test)

# Show the beta coefficients
print('Beta_0: \n', regr.coef_)
print('Beta_1: \n', regr.intercept_)

# Compute the RSS
mse=mean_squared_error(y_test, y_pred)
print('Mean Square Error:',mse)

# Compute the R-square index
rsquare = r2_score(y_test, y_pred)
print('R-square:', rsquare)
```
